# One-Hot Encoder

Applying One-Hot Encoder on the different columns to make the data ready for training and testing

```python
df_final_ohe = pd.get_dummies(data=df_final ,columns=['Total','Age(years)','Education','Gender','Income','Race/Ethnicity'], drop_first=True)
```

# Now applying the SGD Regressor on df_final_ohe

```python
X = df_final_ohe[['YearStart', 'YearEnd', 'LocationDesc', 'Topic', 'Question',
       'Sample_Size', 'QuestionID', 'Stratification1',
       'Age(years)_25 - 34', 'Age(years)_35 - 44', 'Age(years)_45 - 54',
       'Age(years)_55 - 64', 'Age(years)_65 or older',
       'Education_High school graduate', 'Education_Less than high school',
       'Education_Some college or technical school', 'Gender_Male',
       'Income_$25,000 - $34,999', 'Income_$35,000 - $49,999',
       'Income_$50,000 - $74,999', 'Income_$75,000 or greater',
       'Income_Data not reported', 'Income_Less than $15,000',
       'Race/Ethnicity_American Indian/Alaska Native', 'Race/Ethnicity_Asian',
       'Race/Ethnicity_Hawaiian/Pacific Islander', 'Race/Ethnicity_Hispanic',
       'Race/Ethnicity_Non-Hispanic Black',
       'Race/Ethnicity_Non-Hispanic White', 'Race/Ethnicity_Other']]
```

```python
y = df_final_ohe[['Data_Value']]
```

```python
from sklearn import linear_model
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
reg = make_pipeline(StandardScaler(),linear_model.SGDRegressor())
reg.fit(X_train, y_train)
```
