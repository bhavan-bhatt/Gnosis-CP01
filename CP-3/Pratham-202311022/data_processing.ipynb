```python
cols = ['team', 'date', 'event', 'match_number', 'venue', 'city', 'toss_winner', 'toss_decision', 'winner','winner_runs', 'winner_wickets']
for i in os.listdir(path):
    if 'info' in i:
        print(i)
        curr_df = pd.read_csv(path + '/' + i, on_bad_lines='skip')
        curr_df = curr_df.reset_index(drop=True)
        curr_df = curr_df.T
        curr_df = curr_df.reset_index(drop=True)
        curr_df.columns = curr_df.iloc[0]
        curr_df = curr_df.drop(0)
        if 'outcome' in curr_df.columns:
                continue
        for columns in curr_df.columns:
            if columns not in cols:
                try:
                    curr_df = curr_df.drop(columns, axis=1)
                except KeyError:
                    print(i, columns)
        temp = set(cols).symmetric_difference(curr_df.columns)
        diff = list(temp)
        if diff:
            for missing_cols in diff:
                curr_df.insert(loc = cols.index(missing_cols), column = missing_cols, value = '')
        print(curr_df.columns)
        curr_df.columns = ['team_1', 'team_2', 'date', 'event', 'match_number', 'venue', 'city',
                    'toss_winner', 'toss_decision', 'winner','winner_runs', 'winner_wickets']
        pd.concat([curr_df, df])
```
