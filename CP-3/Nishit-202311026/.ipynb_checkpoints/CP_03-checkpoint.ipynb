```python
X = df4.drop(columns=['Match Winner','Match Winner Encoded','Team1 Playing 11', 'Team2 Playing 11'], axis=1)
y = df4['Match Winner Encoded']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
y_test = np.array(y_test)
y_train = np.array(y_train)
```

```python
X_train.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(32, activation='relu', input_shape=(5,)),
#     keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
#Define the model-building function for hyperparameter tuning
from kerastuner.tuners import RandomSearch
def build_model(hp):
    model = Sequential()
    model.add(layers.Dense(units=hp.Int('units1', min_value=32, max_value=128, step=32),
                           activation='relu', input_shape=(X_train.shape[1],)))
    model.add(layers.Dense(units=hp.Int('units2', min_value=32, max_value=128, step=32),
                           activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))
    
    # Compile the model
    model.compile(optimizer=Adam(hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    
    return model

# Initialize the tuner
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,  # adjust based on your resources and requirements
    directory='tuner_directory',  # a directory to save the tuning results
    project_name='classification_tuning'
)

# Perform hyperparameter search
tuner.search(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

# Get the best hyperparameters
best_hp = tuner.get_best_hyperparameters(num_trials=1)[0]

# Build the final model with the best hyperparameters
final_model = tuner.hypermodel.build(best_hp)

# Train the final model
final_model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

# Evaluate the final model on the test set
accuracy = final_model.evaluate(X_test, y_test)[1]
print(f'Test Accuracy: {accuracy}')
```
