# One Hot Encoder 


```python
df_final_ohe = pd.get_dummies(data=df_final ,columns=['YearStart','LocationDesc','Total','Age(years)','Education','Gender','Income','Race/Ethnicity'], drop_first=True)
df_final_ohe
```

```python
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = np.array(df_final_ohe['Age(years)','Education','Gender','Income','Race/Ethnicity','YearStart']).reshape(-1, 1)
y = np.array(df_final_ohe['Data_Value']).reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
regr = LinearRegression()
regr.fit(X_train, y_train)
#print(regr.score(X_test, y_test))
y_pred=regr.predict(X_test)
```
