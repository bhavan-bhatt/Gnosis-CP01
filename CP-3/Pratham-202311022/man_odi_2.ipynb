```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
import os
import re
```

```python
path = 'H:/work_projects/College/DM_Projects/course_project_3/odi_Matches_Data.csv/man_odi_2.csv'
```

```python
df = pd.read_csv(path)
```

```python
df = df.drop(columns=['Unnamed: 0'], axis=1)
```

```python
winning_runs = []
winning_wickets = []
remaining_balls = []
for idx, i in enumerate(df['Match Result Text']):
    try:
        if 'runs' in i:
            winning_runs.append(i)
            winning_wickets.append('')
            remaining_balls.append('')
        elif 'wicket' in i:
            winning_wickets.append(i.split('(', 2)[0])
            remaining_balls.append(i.split('(', 2)[1])
            winning_runs.append('')
        else:
            winning_runs.append('')
            winning_wickets.append('')
            remaining_balls.append('')
    except Exception as e:
        print(e, idx)
```

```python
len(winning_runs), len(winning_wickets), len(remaining_balls)
```

```python
winning_wickets2 = []
for i in winning_wickets:
    if i != '':
        re_wickets = ''.join(re.findall('[0-9]', i))
        if re_wickets != '':
            winning_wickets2.append(int(re_wickets))
        else:
            winning_wickets2.append(np.nan)
        # print(re.findall('[0-9]', i))
    else:
        winning_wickets2.append(np.nan)
```

```python
winning_runs2 = []
for i in winning_runs:
    if i != '':
        re_runs = ''.join(re.findall('[0-9]', i))
        if re_runs != '':
            winning_runs2.append(int(re_runs))
        else:
            winning_runs2.append(np.nan)
        # print(re.findall('[0-9]', i))
    else:
        winning_runs2.append(np.nan)
```

```python
remaining_balls2 = []
for i in remaining_balls:
    if i != '':
        re_balls = ''.join(re.findall('[0-9]', i))
        if re_balls != '':
            remaining_balls2.append(int(re_balls))
        else:
            remaining_balls2.append(np.nan)
        # print(re.findall('[0-9]', i))
    else:
        remaining_balls2.append(np.nan)
```

```python
len(winning_runs2), len(winning_wickets2), len(remaining_balls2)
```

```python
df['winning_wickets'] = winning_wickets2
df['winning_runs'] = winning_runs2
df['remaining_balls'] = remaining_balls2
```

```python
df = df.drop(columns=['Match Result Text'], axis=1)
```
